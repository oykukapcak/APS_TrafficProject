```python
import sumo_runner
import numpy as np
from matplotlib import pyplot as plt
```

```python
normres = sumo_runner.simulate_n_steps(1, 'nogui')
```

```python
gares = sumo_runner.simulate_n_steps(2, 'nogui')
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (8, 6)
N = 4
smooth_norm = np.convolve(normres, np.ones((N,))/N, mode='valid')
smooth_ga = np.convolve(gares, np.ones((N,))/N, mode='valid')
plt.plot(smooth_norm, label='Static')
plt.plot(smooth_ga, label='GA')
plt.legend()
plt.grid()
plt.ylim(ymin=0)
plt.title("Comparison of #waiting cars (Dynamic flow)")
plt.ylabel('#Waiting cars')
plt.xlabel("Phase cycles (1 is 120 ticks)")
plt.show()
```
